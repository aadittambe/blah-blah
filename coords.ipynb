```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import time
```

```python
data = pd.read_csv("../../data/processed/etl_schools_all_years.csv")
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('.', '_')
data['district_x'] = data['district_x'].str.replace("\n"," ")
data['school_x'] = data['school_x'].str.replace("\n"," ")
data['street'] = data['street'].str.replace("\n"," ")

data
```

```python
data["lat"] = ""
data["lon"] = ""
```

```python
counter = 0
for index,row in data.iterrows():
    address = f"{row['street']}, {row['city']}, {row['state']} {row['zip']}"
    # print(address)
    geolocator = Nominatim(user_agent="Aadit Tambe, University of Maryland")
    try:
        counter = counter + 1
        location = geolocator.geocode(address)
        data["lat"][index] = location.latitude
        data["lon"][index] = location.longitude
        print(str(counter) + ': done')
        time.sleep(2)
    except:
        counter = counter + 1
        data["lat"][index] = "no_result"
        data["lon"][index] = "no_result"
        print(str(counter) +': no result')
        time.sleep(3)
```
